## # Find employees who have worked for 7 consecutive days

```python
from datetime import timedelta
```

### The NAME and POSITION of employees 
###      A) who has worked for 7 consecutive days.

```python
df2['Time'] = pd.to_datetime(df2['Time'], errors='coerce')

# Sort the DataFrame by 'Employee Name' and 'Time'
df2 = df2.sort_values(by=['Employee Name', 'Time'])

# Calculate the day difference for each employee
df['Day Difference'] = df.groupby('Employee Name')['Time'].diff().dt.days.fillna(0).astype(int)

# Find employees who have worked for 7 consecutive days

consecutive_days = 7
result_df = df2[df2['Day Difference'].eq(consecutive_days - 1)]


from tabulate import tabulate
print("Employees who worked for 7 consecutive days:")
table = tabulate(result_df[['Employee Name', 'Position ID']], headers='keys', tablefmt='pretty', showindex=False)
print(table)
```

```python
# Check for the time difference between shifts
df2['Time Difference'] = df2['Time'].shift(-1) - df2['Time Out']
```

```python
# Filter entries with time difference between 1 and 10 hours
filtered_df = df[(df['Time Difference'] > pd.Timedelta(hours=1)) & (df['Time Difference'] < pd.Timedelta(hours=10))]
```

### The NAME and POSITION of employees

```python
# Display the result
filtered_df[['Employee Name', 'Position ID']].drop_duplicates()
```

```python
df2['Time-In'] = pd.to_datetime(df2['Time-In'])
df2['Time-Out'] = pd.to_datetime(df2['Time-Out'])

# Calculate the duration of each shift
df2['Shift_Duration'] = df2['Time-Out'] - df2['Time-In']

# Filter employees who worked for more than 14 hours in a single shift
fc_df = df2[df2['Shift_Duration'] > '14:00:00']
```
